```python
import pandas as pd
import geopandas as gpd
from shapely.geometry import Point
```

```python
# Load the CSV file (Global Coastal Dataset)
csv_file_path = r"C:\TU_Delft\Internship\Execution\COMPASS-SFINCS_Wave\SFINCS_Wave\boundary_conditions\GCC\GCC_geophysical.csv"

# Load the dataset
gcc_df = pd.read_csv(csv_file_path)

# Drop rows where 'ns' has NaN values, but keep rows with valid 'lon' and 'lat'
gcc_df = gcc_df[['lon', 'lat', 'ns', 'bs_deltadtm', 'cs_deltadtm']].dropna(subset=['ns'])
gcc_gdf = gpd.GeoDataFrame(gcc_df, geometry=gpd.points_from_xy(gcc_df['lon'], gcc_df['lat']), crs="EPSG:4326")
```

```python
gcc_gdf
```

```python
obs_coord = [28.935,-95.295]
obs_point = Point(obs_coord[1], obs_coord[0])  # (lon, lat)

# Calculate distances to obs_point
gcc_gdf['distance'] = gcc_gdf['geometry'].distance(obs_point)

# Find the row with the minimum distance
closest_point_df = gcc_gdf.loc[[gcc_gdf['distance'].idxmin()]]
dataset_path = r"./Clean/beach_slope.csv"
closest_point_df.to_csv(dataset_path, index=True)

# Display the DataFrame
print("Closest point as a DataFrame:")
print(closest_point_df)
```
